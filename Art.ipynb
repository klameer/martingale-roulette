```python
# Functions

from itertools import cycle
import turtle
import random 

def draw_door(t, x=0, y=0, scale=1, color='blue', text='', style=('Georgia', 10, 'bold')):
    t.home()
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.pensize(2)
    t.forward(50 * scale)
    t.left(90)

    #knob
    t.forward(40 * scale)
    t.left(90)
    t.forward(10 * scale)
    t.right(90)
    t.fillcolor(color)
    t.begin_fill()
    t.circle(5 * scale)
    t.end_fill()
    t.left(90)
    t.backward(10 * scale)

    t.right(90)
    t.forward(40 * scale)
    t.left(90)
    t.forward(50 * scale)
    t.left(90)
    t.forward(80 * scale)
    t.left(90)
    t.penup()
    t.forward(10)
    t.left(90)
    t.forward(50 * scale)
    t.write(text, font=style)
    t.home()
    return(t)

# Draw Guy
def draw_guy(t, x=0, y=0, scale=1, color='blue', text='', style=('Georgia', 10, 'bold')):
    
    t.hideturtle()
    t.home()
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.pensize(2)
    t.right(90)
    t.forward(20 * scale)
    t.left(45)
    t.forward(10 * scale)
    t.backward(10 * scale)
    t.right(45)
    t.right(45)
    t.forward(10 * scale)
    t.backward(10 * scale)
    t.left(45)
    t.backward(15 * scale)
    t.left(45)
    t.forward(10 * scale)
    t.backward(10 * scale)
    t.left(90)
    t.backward(10 * scale)
    t.forward(10 * scale)
    t.left(45)
    t.forward(5 * scale)
    t.fillcolor(color)
    t.right(90)
    t.begin_fill()
    t.circle(10 * scale)
    t.end_fill()
    t.penup()
    t.forward(10 * scale)
    t.left(90)
    t.forward(10 * scale)
    t.pendown()
    t.write(text, font=style)

    t.penup()
    t.home()
    t.pendown()
    
    return(t)
    
    
def draw_roulette(t, x=0, y=0, scale=1, text='', style=('Georgia', 10, 'bold')):
    t.home()
    t.penup()
    t.goto(x, y)
    t.pendown()
    f = cycle(['red', 'black'])

    for i in range(38):    
        if (i == 37) | (i == 19):
            t.fillcolor('green')
        else:
            t.fillcolor(next(f))

        t.circle(30 * scale,360/38)

        t.begin_fill()
        t.right(90)
        t.forward(10 * scale)
        t.right(90)
        t.forward(10 * scale)
        t.right(90)
        t.forward(10 * scale)
        t.right(90)
        t.forward(10 * scale)

        t.end_fill()

    t.left(90)
    t.penup()
    t.forward(25 * scale)
    t.right(90)
    t.pendown()

    t.fillcolor('black')
    t.begin_fill()
    t.circle(5 * scale)
    t.end_fill()  

    t.left(90)
    t.penup()
    t.forward(50 * scale)
    t.pendown()
    t.write(text, font=style)
    t.penup()
    t.home()
    t.pendown()
    
    return(t)
```

```python
# Draw Martin and Table
screen = turtle.Screen()
screen.setup(500,500)
screen.reset()

t = turtle.Turtle()

t.reset()
t.speed(0)

t = draw_guy(t, x=-100, scale=3, text='Martin')
t = draw_roulette(t, x=100, y=-50, text="Table")

for k in turtle.turtles():
    k.hideturtle()

cv = turtle.getcanvas()
cv.postscript(file='martin_table.eps')

turtle.done()
```

```python
# Draw Many Martin and Table
screen = turtle.Screen()
screen.setup(500,500)
screen.reset()

t = turtle.Turtle()

t.reset()
t.speed(0)

for i in range(100):
    x = random.randint(-250,0)
    y = random.randint(-250, 250)

    t = draw_guy(t, x=x, y=y, scale=2, text='')
    
t = draw_roulette(t, x=100, y=-50, text="Table")

for k in turtle.turtles():
    k.hideturtle()

cv = turtle.getcanvas()
cv.postscript(file='martin_table_many.eps')

turtle.done()
```
